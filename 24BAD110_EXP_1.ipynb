```
# Scenario 1

print("Name: Sharvesh S")
print("Roll No: 24BAD110")
print("Scenario 1: E-commerce Sales Data Analysis\n")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data.csv", encoding='ISO-8859-1')

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())

print("Missing Values:\n", df.isnull().sum())

df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

sales_product = df.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)

plt.figure()
sales_product.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.xticks(rotation=90)
plt.show()

plt.figure()
sales_product.plot(kind='line')
plt.title("Top 10 Products Sales Trend")
plt.show()
```

```
# Scenario 2

print("Name: Sharvesh S")
print("Roll No: 24BAD110")
print("Scenario 2: Hospital Patient Records Analysis\n")

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("diabetes.csv")

print(df.head())
print(df.info())
print(df.describe())

print("Missing Values:\n", df.isnull().sum())

plt.figure()
plt.hist(df['Glucose'], bins=20)
plt.title("Glucose Level Distribution")
plt.show()

plt.figure()
plt.boxplot(df['Age'])
plt.title("Age Distribution")
plt.show()

plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Between Health Metrics")
plt.show()
```

```

print("Name: Sharvesh S")
print("Roll No: 24BAD110")
print("Scenario 3: Housing Dataset Analysis\n")

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Housing.csv")

# Display basic info
print(df.head())
print("\nDataset Info:\n")
print(df.info())

# Missing values
print("\nMissing Values:\n", df.isnull().sum())

# Scatter Plot (Area vs Price)
plt.figure()
plt.scatter(df['area'], df['price'])
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()

# â Convert categorical columns into numeric using encoding
df_encoded = df.copy()

for col in df_encoded.select_dtypes(include='object').columns:
    df_encoded[col] = df_encoded[col].astype('category').cat.codes

# Heatmap after encoding
plt.figure(figsize=(10,8))
sns.heatmap(df_encoded.corr(), annot=True)
plt.title("Feature Correlation Heatmap (Encoded Data)")
plt.show()
```

```
# scenario 4
5
print("Name: Sharvesh S")
print("Roll No: 24BAD110")
print("Scenario 4: Banking Customer Data Analysis\n")

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# â IMPORTANT FIX: use tab separator
df = pd.read_csv("marketing_campaign.csv", sep="\t")

# First 5 rows
print("First 5 Rows:\n")
print(df.head())

# Dataset info
print("\nDataset Info:\n")
print(df.info())

# Summary statistics
print("\nStatistical Summary:\n")
print(df.describe())

# Missing values
print("\nMissing Values:\n")
print(df.isnull().sum())

# -----------------------------
# Create Age column
# -----------------------------
df['Age'] = 2026 - df['Year_Birth']

plt.figure()
plt.hist(df['Age'], bins=20)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# -----------------------------
# Income Boxplot
# -----------------------------
plt.figure()
plt.boxplot(df['Income'].dropna())
plt.title("Income Distribution")
plt.ylabel("Income")
plt.show()

# -----------------------------
# Correlation Heatmap
# -----------------------------
numerical_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(12,10))
sns.heatmap(numerical_df.corr(), annot=True)
plt.title("Correlation Between Numerical Features")
plt.show()
```
